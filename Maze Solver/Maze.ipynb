##Maze Output Function

```python
def output_maze(grid, is_solving):
	
	# Adding New Line for Output Cleanliness
	print()
 
 	# Looping Through Every Row 
	for i in grid:
		
		# Looping Through Every Column
		for j in i:
			
			# Checking Current Value
			if j == "A":
				
				# Checking Solving State
				if is_solving:
					print("\033[1;31mA\033[0;0m", end = "")
		 
		 		# Checking Solving State
				else:
					print("\033[5;31mA\033[0;0m", end = "")
		 
		 	# Checking Current Value
			elif j == "B":
				
				# Checking Solving State
				if is_solving:
					print("\033[5;32mB\033[0;0m", end = "")
		 
		 		# Checking Solving State
				else:
					print("\033[1;32mB\033[0;0m", end = "")
		 
		 	# Checking Current Value
			elif j == "*":
				print("\033[1;36m*\033[0;0m", end = "")
		
			# Checking Current Value
			elif j == "#":
				print("\033[1;40;40m \033[0;0m", end = "")
		
			# Checking Current Value
			else:
				print(" ", end = "")
		
		# Adding New Line for Output Cleanliness
		print()
```

##Parsing String Input

```python
# Initializing Column Length Variable
col_length = 0

# Creating While Loop to Count # of Columns
while grid[col_length] != "\n":
  
  # Increasing Column Length Variable
	col_length += 1

# Finding Row Length Using Mod Operator
row_length = len(grid) % col_length

# Creating 2D Maze Array Using List Comprehension
maze = [[""]*col_length for i in range(row_length)]

# Declaring Count Variable
count = 0

# Increasing Column Length
col_length += 1

# Using For Loop to Go Through Maze String (Grid)
for i in range(len(grid)):
  
  # Checking for End of String
	if grid[i] != "\n":
    
    # Filling in 2D Array
		maze[count // col_length][count % col_length] = grid[i]
    
  # Increasing Count Variable
  count += 1

# Outputting Maze Using Function
output_maze(maze, False)
```

##Creating Adjacency List

```python
# Declaring Adjacency Dictionary
adjacency = {}

# Declaring Start/End Coordinates (AKA A & B)
start = [0,0]
end = [0,0]

# Looping Through Every Row in 2D Array
for i in range(len(maze)):
	
	# Looping Through Every Column
	for j in range(len(maze[i])):
		
		# Checking for Blank Space or Start Position
		if maze[i][j] == " " or maze[i][j] == "A":
			
			# Checking for Start Position
			if maze[i][j] == "A":
				
				# Storing Start Position
				start = [i,j]
				
			# Creating Index String to Track Position in Dictionary
			index = f"{i},{j}"
			
			# Creating Empty Array in Adjacency Dictionary
			adjacency[index] = []

			try:
				
				# Checking North Position for Empty Space or End Position
				if i-1 >= 0 and (maze[i-1][j] == " " or maze[i-1][j] == "B"):
					
					# Adding Position to Dictionary Array
					adjacency[index].append([i-1,j])
     
			except: pass

			try:
				
				# Checking West Position for Empty Space or End Position
				if j-1 >= 0 and (maze[i][j-1] == " " or maze[i][j-1] == "B"):
					
					# Adding Position to Dictionary Array
					adjacency[index].append([i,j-1])
     
			except: pass

			try:
				
				# Checking South Position for Empty Space or End Position
				if maze[i+1][j] == " " or maze[i+1][j] == "B":
					
					# Adding Position to Dictionary Array
					adjacency[index].append([i+1,j])
          
			except: pass

			try:
				
				# Checking East Position for Empty Space or End Position
				if maze[i][j+1] == " " or maze[i][j+1] == "B":
					
					# Adding Position to Dictionary Array
					adjacency[index].append([i,j+1])
     
			except: pass
			
		# Checking for End Position
		elif maze[i][j] == "B":
			
			# Storing End Position
			end = [i,j]
```

##Defining Search Algorithm

```python
# Defining BFS Algorithm
def bfs(start, end, map):
  
  # Declaring Blank Visited Array
  visited = []
  
  # Declaring Queue Array
  queue = [(start,[start])]
  
  # Creating a While Loop to Go Through the Queue 
  while len(queue) > 0:
    
    # Accessing First Move in Queue
    vertex, path = queue.pop(0)
    
    # Adding Location to Visited Array
    visited.append(vertex)
    
    # Creating Index String Using Coordinates
    index = f"{vertex[0]},{vertex[1]}"
    
    # Looping Through Every Connection in Adjacency List
    for connection in map[index]:
      
      # Checking for End Location
      if [connection[0], connection[1]] == end:
        
        # Returning Path
        return path + [end]
        
      # When Not End Location
      else:
        
        # Checking For New Connection
        if connection not in visited:
          
          # Adding Connection to Visited Array
          visited.append(connection)
          
          # Adding Connection & Modified Path to Queue
    			queue.append((connection, path + [connection]))
       
  # Returning False for Negative Result
  return False
```

##Searching For Path & Completing Maze

```python
# Calling BFS Algorithm to Complete Maze
search_result = bfs(start, end, adjacency)

# Checking Search Result
if search_result == False:
  
  # Outputting Error Message
  print("\nERROR! No path could be found. :(")
 
else:
  
  # Outputting Success Message
  print("\nSOLUTION FOUND!")
  
  # Adding Stars to Indicate Path
  for i in search_result[1:-1]:
    
    # Setting Grid Location to Star
    maze[i[0]][i[1]] = "*"
    
# Outputting Solved Form of Maze
output_maze(maze, True)
```
